# Linux (Debian) – práce se soubory v shellu

Tento notebook slouží jako **studijní materiál pro předmět Operační systémy**.

Zaměřuje se na **základní příkazy Linuxového shellu (Debian/Ubuntu)** pro práci se soubory:
- vytváření souborů
- kopírování
- přesun a přejmenování
- mazání
- editaci a přidávání obsahu
- práci s odkazy (linky)

U každého příkazu je:
- stručný popis (Markdown)
- alespoň jedna praktická ukázka v `bash`.

## 1. `touch`

Vytvoří prázdný soubor nebo aktualizuje čas poslední změny existujícího souboru.

```
touch soubor.txt
```

## 2. `ls`

Vypíše obsah adresáře.

```
ls -l
```

## 3. `cp`

Kopíruje soubory nebo adresáře.

```
cp soubor.txt kopie_souboru.txt
```

## 4. `mv`

Přesouvá nebo přejmenovává soubory a adresáře.

```
mv kopie_souboru.txt archiv_souboru.txt
```

## 5. `rm`

Maže soubory nebo adresáře.

```
rm archiv_souboru.txt
```

## 6. `cat`

Zobrazí obsah souboru nebo spojí více souborů.

```
cat soubor.txt
```

## 7. `echo` + přesměrování

Zapíše text do souboru.

- `>` přepíše obsah souboru
- `>>` přidá obsah na konec souboru (append)

```
echo "První řádek" > soubor.txt
echo "Další řádek" >> soubor.txt
```

## 8. `nano`

Jednoduchý terminálový textový editor vhodný pro začátečníky.

```
nano soubor.txt
```

## 9. `ln`

Vytváří odkazy (linky).

- bez přepínače: hard link
- `-s`: symbolický (soft) link

```
ln soubor.txt hardlink.txt
ln -s soubor.txt softlink.txt
```

## 10. `mkdir`

Vytváří adresáře.

```
mkdir data
```
